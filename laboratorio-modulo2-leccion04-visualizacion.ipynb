```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix_titles =pd.read_csv("datos/netflix_titles.csv")
netflix_titles.drop(columns = ["show_id","Unnamed: 0"], inplace=True)
netflix_titles
```

```python
netflix_original = pd.read_csv("datos/netflix_originals.csv")
netflix_original.drop(columns = ["Unnamed: 0"], inplace=True)
netflix_original
```

```python
netflix_original.rename(columns={'Title': 'title'}, inplace=True)

df= pd.merge(netflix_titles, netflix_original, on=["title"], how='outer')
df
```

#### Ejercicio 10: ¿Cómo se distribuyen los títulos por país para los 6 principales países? Tienes que crear una gráfica para cada país mostrando los resultados en un subplot. Deberás:

#### Ejercicio 11: ¿Cuál es total de tipos de calificaciones, considerando solo los principales 5 países? Deberás:

```python
top_5 = df.groupby("country").count().nlargest(5, columns="title").reset_index()
lista_paises = list(top_5["country"])
lista_paises
```

#### Ejercicio 12: ¿Cómo se distribuyen las 5 primeras calificaciones a lo largo de los últimos 10 años? 
#### (Las 5 primeras calificaciones son las que tienen mas titulos) [grafico de lineas una por cada rating]

- Poner nombre a la ejes. 

- Poner título a la gráfica. 

- Rota las etiquetas del eje X. 

- Elimina el nombre del eje X. 

```python
top5 = df['rating'].value_counts()[:5].index.to_list()
top5
```

```python
filtro_rating = df["rating"].isin(top5)
anio_max = df["release_year"].unique()[0]
print(anio_max)
filtro_años = df["release_year"].between(anio_max-10,anio_max,inclusive="both")
netflix_filtrado = df[ filtro_rating & filtro_años ]
netflix_filtrado
```

```python
plt.figure(figsize=(15,5))
plt.title("Calificaciones distintas por top 5 paises últimos 10 años")
sns.lineplot(data=netflix_filtrado, 
            x="release_year",
            y="title",
            hue="rating",
            estimator="sum",
            palette="hsv")
plt.ylabel("Titulos")
plt.xlabel("Año")
plt.show()
```
